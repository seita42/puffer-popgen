### PCA analysis with EIGENSOFT 7.2.1

```python
col_Mj = "lightskyblue"
col_Mp = "lightseagreen"
col_Kj = "khaki"
col_Kp = "orange"

mar_Mj = '^'
mar_Mp = '^'
mar_Kj = 'o'
mar_Kp = 'o'
```

```python
df_evec = pd.read_table("pca/MK28_pruned.evec",
                        header=None,
                        skiprows=1,
                        sep='\\s+',
                        skipinitialspace=True)
df_evec.rename(columns={0:"ind",
                        1:"PC1",
                        2:"PC2",
                        3:"PC3",
                        4:"PC4",
                        5:"PC5",
                        6:"PC6",
                        11:"pop"}, inplace=True)
df_evec["ind"] = df_evec["ind"].str.split(":", expand=True)[0]

df_evec["color"] = "black"
df_evec.loc[df_evec['pop']=="Mj", 'color'] = col_Mj
df_evec.loc[df_evec['pop']=="Mp", 'color'] = col_Mp
df_evec.loc[df_evec['pop']=="Kj", 'color'] = col_Kj
df_evec.loc[df_evec['pop']=="Kp", 'color'] = col_Kp

df_evec["marker"] = "o"
df_evec.loc[df_evec['pop']=="Mj", 'marker'] = mar_Mj
df_evec.loc[df_evec['pop']=="Mp", 'marker'] = mar_Mp
df_evec.loc[df_evec['pop']=="Kj", 'marker'] = mar_Kj
df_evec.loc[df_evec['pop']=="Kp", 'marker'] = mar_Kp

df_evec.head()
```

```python
df_eval = pd.read_table("pca/MK28_pruned.eval",
                        header=None,
                        names=["eval"])
df_eval["PC"] = ["PC" + str(x + 1) for x in list(df_eval.index)]
df_eval["proportion"] = df_eval["eval"] / sum(df_eval["eval"])
df_eval.set_index("PC", inplace=True)
df_eval.head()
```

```python
pc1 = df_eval.at['PC1', 'proportion']
pc2 = df_eval.at['PC2', 'proportion']
pc3 = df_eval.at['PC3', 'proportion']
pc4 = df_eval.at['PC4', 'proportion']
pc5 = df_eval.at['PC5', 'proportion']
pc6 = df_eval.at['PC6', 'proportion']

print(pc1, pc2, pc3, pc4, pc5, pc6)
```

```python
# sns.set_context('talk')
# sns.set_style("whitegrid", {'grid.linestyle': '--'})
# sns.set_style("white")
sns.set(style="whitegrid", context="talk", rc={'grid.linestyle': '--'})

def pca_plot(pcx, pcy):

    plt.figure(figsize=(5, 5))
    
    for idx, row in df_evec.iterrows():
        pop = row['pop']
#        label = "$\it{"+row['ind']+"}$"
        label = row['pop']
        color=row['color']
        marker=row['marker']
#        markersize=row['markersize']
#        h_align=row['h_align']
#        alpha=row['alpha']
#        linewidth=row['linewidth']
#        xytext=(row['xx'], row['yy'])
            
        plt_sc = plt.scatter(row[pcx], row[pcy],
                             color=color,
                             marker=marker,
                             s=350,
#                             linewidth=linewidth,
                             alpha=0.8,
                             label=label)

#         plt.annotate(label,
#                      xy=(row[pcx], row[pcy]),
# #                      xytext=xytext,
# #                      textcoords='offset points',
# #                      horizontalalignment=h_align,
# #                      verticalalignment='bottom',
#                      alpha=0.8,
#                      fontsize=10)

    return True

pc1 = df_eval.at['PC1', 'proportion']
pc2 = df_eval.at['PC2', 'proportion']
pc3 = df_eval.at['PC3', 'proportion']
pc4 = df_eval.at['PC4', 'proportion']
pc5 = df_eval.at['PC5', 'proportion']
pc6 = df_eval.at['PC5', 'proportion']

# PC1, PC2
pca_plot('PC1', 'PC2')

plt.xlabel("PC1 (" + "{0:.1f}".format(pc1*100) + "%)", fontsize=24)
plt.ylabel("PC2 (" + "{0:.1f}".format(pc2*100) + "%)", fontsize=24)
plt.xlim(-0.25, 0.25)
plt.ylim(-0.38, 0.38)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# plt.savefig("MK28_PCA_PC1-PC2.svg", bbox_inches='tight')
# plt.savefig("MK28_PCA_PC1-PC2.pdf", bbox_inches='tight')


# PC2, PC3
pca_plot('PC2', 'PC3')

plt.xlabel("PC2 (" + "{0:.1f}".format(pc2*100) + "%)", fontsize=24)
plt.ylabel("PC3 (" + "{0:.1f}".format(pc3*100) + "%)", fontsize=24)
plt.xlim(-0.35, 0.35)
plt.ylim(-0.35, 0.68)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# plt.savefig("MK28_PCA_PC2-PC3.svg", bbox_inches='tight')
# plt.savefig("MK28_PCA_PC2-PC3.pdf", bbox_inches='tight')

# PC3, PC4
pca_plot('PC3', 'PC4')

plt.xlabel("PC3 (" + "{0:.1f}".format(pc3*100) + "%)", fontsize=24)
plt.ylabel("PC4 (" + "{0:.1f}".format(pc4*100) + "%)", fontsize=24)
plt.xlim(-0.35, 0.68)
plt.ylim(-0.95, 0.9)
plt.xticks(np.arange(-0.2, 0.7, 0.2), fontsize=12)
plt.yticks(np.arange(-0.8, 0.9, 0.4), fontsize=12)

# plt.savefig("MK28_PCA_PC3-PC4.svg", bbox_inches='tight')
# plt.savefig("MK28_PCA_PC3-PC4.pdf", bbox_inches='tight')

# PC4, PC5
pca_plot('PC4', 'PC5')

plt.xlabel("PC4 (" + "{0:.1f}".format(pc4*100) + "%)", fontsize=24)
plt.ylabel("PC5 (" + "{0:.1f}".format(pc5*100) + "%)", fontsize=24)
plt.xlim(-0.95, 0.85)
plt.ylim(-0.5, 1.1)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# plt.savefig("MK28_PCA_PC4-PC5.svg", bbox_inches='tight')
# plt.savefig("MK28_PCA_PC4-PC5.pdf", bbox_inches='tight')

# PC5, PC6
pca_plot('PC5', 'PC6')

plt.xlabel("PC5 (" + "{0:.1f}".format(pc5*100) + "%)", fontsize=24)
plt.ylabel("PC6 (" + "{0:.1f}".format(pc6*100) + "%)", fontsize=24)
plt.xlim(-0.6, 1.1)
plt.ylim(-0.8, 0.4)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# plt.savefig("MK28_PCA_PC5-PC6.svg", bbox_inches='tight')
# plt.savefig("MK28_PCA_PC5-PC6.pdf", bbox_inches='tight')

plt.show()
```

### ADMIXTURE analysis with ADMIXTURE 1.3.0

```python
cvs = []

for K in range(1, 9):
    log_file = f"admixture/cv/log_{K}.out"
    with open(log_file, "r") as f:
        for line in f:
            match = re.search(r"CV error.*?:\s+([\d\.]+)", line)
            if match:
                cvs.append(float(match.group(1)))
                break
Ks = list(range(1, 9))

print("K values:", Ks)
print("CV errors:", cvs)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(Ks[0:5], cvs[0:5], marker='o', linewidth=3)
# plt.xlabel("K", fontsize=24)
plt.text(x= 0.7, y= 0.4467, s="K =", fontsize=20, ha='right', va='top')
plt.ylabel("CV error", fontsize=24)
plt.ylim(0.46, 0.7)
plt.xticks(Ks[0:5], fontsize=20)
plt.yticks(fontsize=16)
plt.tight_layout()

# plt.savefig("MK28_ADMIXTURE_CV_error.pdf", bbox_inches='tight')
# plt.savefig("MK28_ADMIXTURE_CV_error.svg", bbox_inches='tight')

plt.show()
```

```python
sns.set_context("talk")
sns.set_style("white")

basename = "admixture/MK28_pruned"

sample_order=['Mj003',
              'Mj006',
              'Mj065',
              'Mj098',
              'Mj101',
              'Mj117',
              'Mj164',
              'Mp013',
              'Mp014',
              'Mp062',
              'Mp063',
              'Mp114',
              'Mp205',
              'Mp208',
              'Kj020',
              'Kj026',
              'Kj067',
              'Kj094',
              'Kj139',
              'Kj166',
              'Kj066',
              'Kp179',
              'Kp182',
              'Kp019',
              'Kp136',
              'Kp180',
              'Kp190',
              'Kp194']

for i in [2, 3, 4]:
    plt.figure(figsize=(10, 2.0))
    df_k = pd.read_csv(basename+"."+str(i)+".Q",
                       header=None,
                       sep='\\s+')
    df_k.columns = list(range(1, i+1))
    df_k['sample_name'] = pd.Categorical(smnames[0][0:28], sample_order)
    # df_k['sample_name'] = pd.Categorical(df_k['sample_name'], sample_order)
    df_k['label'] = df_k['sample_name']

    df_k_sort=df_k.sort_values(by='sample_name')
   
    r = np.arange(len(df_k_sort))
    
    if (i==2):
        plt.bar(r, list(df_k_sort[2]), 0.8, linewidth=0, color=col_Mp, align='center')
        plt.bar(r, list(df_k_sort[1]), 0.8, linewidth=0, color=col_Kp,
                bottom=list(df_k_sort[2]))
    elif (i==3):
        plt.bar(r, list(df_k_sort[2]), 0.8, linewidth=0, color=col_Mj, align='center')
        plt.bar(r, list(df_k_sort[1]), 0.8, linewidth=0, color=col_Mp,
                bottom=list(df_k_sort[2]))
        plt.bar(r, list(df_k_sort[3]), 0.8, linewidth=0, color=col_Kp,
                bottom=list(df_k_sort[1]+df_k_sort[2]))
    elif (i==4):
        plt.bar(r, list(df_k_sort[1]), 0.8, linewidth=0, color=col_Mj, align='center')
        plt.bar(r, list(df_k_sort[4]), 0.8, linewidth=0, color=col_Kp,
                bottom=list(df_k_sort[1]))
        plt.bar(r, list(df_k_sort[3]), 0.8, linewidth=0, color=col_Mp,
                bottom=list(df_k_sort[1]+df_k_sort[4]))
        plt.bar(r, list(df_k_sort[2]), 0.8, linewidth=0, color=col_Kj,
                bottom=list(df_k_sort[1]+df_k_sort[4]+df_k_sort[3]))
    else:
        plt.bar(r, list(df_k_sort[1]), 0.8)
        bottom = df_k_sort[1]
        for j in np.arange(2, i+1):
            plt.bar(r, list(df_k_sort[j]), 0.8, bottom=list(bottom))
            bottom=bottom+df_k_sort[j]
    
    plt.tick_params(axis='x',
                    which='both',
                    bottom=False,
                    top=False,
                    labelbottom=False)

    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)

    plt.xlim(-0.8, len(df_k_sort)-0.5)
    plt.ylabel("K="+str(i), fontsize=24)
    plt.yticks((0.0, 0.5, 1.0), fontsize=12)

    # plt.savefig("MK28_admixture_K"+str(i)+".svg", bbox_inches='tight')
    # plt.savefig("MK28_admixture_K"+str(i)+".pdf", bbox_inches='tight')

plt.xticks(r, df_k_sort['label'], fontsize=16, rotation=90)
plt.tick_params(
    axis='x',
    labelbottom=True)

# plt.savefig("MK28_admixture_K4.svg", bbox_inches='tight')
# plt.savefig("MK28_admixture_K4.pdf", bbox_inches='tight')
plt.show()
```
