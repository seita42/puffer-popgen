#### Parameters
- **μ**: 5.97×10⁻⁹ (fish average mutation rate per generation; Bergeron et al. 2023 *Nature*: [DOI: 10.1038/s41586-023-05752-y](https://www.nature.com/articles/s41586-023-05752-y))
- **Generation time**: 2.5 years (male: 2 years, female: 3 years)


```python
mu = 5.97e-9
gen = 2.5
```

```python
def calc_t_years(df):
    df['t_years'] = gen * ((df.left_time_boundary + df.right_time_boundary) / 2.0) / mu
    df['p_size'] = (1.0 / df['lambda']) / (2.0 * mu)

def calc_t_all():
    calc_t_years(Mj003)
    calc_t_years(Mj006)
    calc_t_years(Mj065)
    calc_t_years(Mj098)
    calc_t_years(Mj101)
    calc_t_years(Mj117)
    calc_t_years(Mj164)

    calc_t_years(Mp013)
    calc_t_years(Mp014)
    calc_t_years(Mp062)
    calc_t_years(Mp063)
    calc_t_years(Mp114)
    calc_t_years(Mp205)
    calc_t_years(Mp208)

    calc_t_years(Kj020)
    calc_t_years(Kj026)
    calc_t_years(Kj066)
    calc_t_years(Kj067)
    calc_t_years(Kj094)
    calc_t_years(Kj139)
    calc_t_years(Kj166)

    calc_t_years(Kp019)
    calc_t_years(Kp136)
    calc_t_years(Kp179)
    calc_t_years(Kp180)
    calc_t_years(Kp182)
    calc_t_years(Kp190)
    calc_t_years(Kp194)
```

```python
sns.set_context('talk')
sns.set_style("ticks")

col_Mj = "lightskyblue"
col_Mp = "lightseagreen"
col_Kj = "khaki"
col_Kp = "orange"

calc_t_all()

def plot_line(df, color, label):
    plt.loglog(df.t_years, df.p_size, drawstyle='steps', color=color, label=label, alpha=0.6, lw=3)
    
plt.figure(figsize=(9, 6))

plot_line(Mj003, col_Mj, "Mj")
plot_line(Mj006, col_Mj, "Mj")
plot_line(Mj065, col_Mj, "Mj")
plot_line(Mj098, col_Mj, "Mj")
plot_line(Mj101, col_Mj, "Mj")
plot_line(Mj117, col_Mj, "Mj")
plot_line(Mj164, col_Mj, "Mj")

plot_line(Mp013, col_Mp, "Mp")
plot_line(Mp014, col_Mp, "Mp")
plot_line(Mp062, col_Mp, "Mp")
plot_line(Mp063, col_Mp, "Mp")
plot_line(Mp114, col_Mp, "Mp")
plot_line(Mp205, col_Mp, "Mp")
plot_line(Mp208, col_Mp, "Mp")

plot_line(Kj020, col_Kj, "Kj")
plot_line(Kj026, col_Kj, "Kj")
plot_line(Kj066, col_Kj, "Kj")
plot_line(Kj067, col_Kj, "Kj")
plot_line(Kj094, col_Kj, "Kj")
plot_line(Kj139, col_Kj, "Kj")
plot_line(Kj166, col_Kj, "Kj")

plot_line(Kp019, col_Kp, "Kp")
plot_line(Kp136, col_Kp, "Kp")
plot_line(Kp179, col_Kp, "Kp")
plot_line(Kp180, col_Kp, "Kp")
plot_line(Kp182, col_Kp, "Kp")
plot_line(Kp190, col_Kp, "Kp")
plot_line(Kp194, col_Kp, "Kp")

plt.xlabel("Years before the present", fontsize=24)
plt.ylabel("Effective population size", fontsize=24)
plt.ylim(1, 300000)
plt.xlim(3000,)

plt.ylim(8e3, 1e6)
plt.xlim(3000,9.0e6)

ax = plt.gca()
formatter = LogFormatterMathtext(base=10)
ax.yaxis.set_major_formatter(formatter)
# plt.tick_params(axis='both', which='major', labelsize=16)

legend_elements = [Line2D([0], [0], color=col_Mj, lw=8, alpha=1.0),
                   Line2D([0], [0], color=col_Mp, lw=8, alpha=1.0),
                   Line2D([0], [0], color=col_Kj, lw=8, alpha=1.0),
                   Line2D([0], [0], color=col_Kp, lw=8, alpha=1.0)]
ax.legend(legend_elements, ['Mj', 'Mp', 'Kj', 'Kp'], fontsize=18,
          frameon=False, loc='best')

# plt.savefig('msmc_gen_2.5_mu_5.97e-9_LogLog.svg')
# plt.savefig('msmc_gen_2.5_mu_5.97e-9_LogLog.pdf')
plt.show()
```

#### Divergence time estimation based on rCCR (Relative Cross-Coalescence Rate)
- rCCR was calculated for all 196 pairwise combinations among 14 individuals from each species
- Read-based phasing of variants was conducted using **WhatsHap** prior to MSMC2 analysis.

```python
sns.set_context('talk')
sns.set_style("ticks")

files = glob.glob("msmc/combined_*.final.txt")

rccr_list = []
time_list = []

for f in files:
    df = pd.read_csv(f, sep="\t")
    time = (df["left_time_boundary"] + df["right_time_boundary"]) / 2
    real_time = time / mu * gen

    rccr = df["lambda_01"] / ((df["lambda_00"] + df["lambda_11"]) / 2)

    time_list.append(real_time.values)
    rccr_list.append(rccr.values)

rccr_array = np.vstack(rccr_list)
rccr_mean = rccr_array.mean(axis=0)
rccr_std = rccr_array.std(axis=0)

rep_df = pd.read_csv(files[0], sep="\t")
rep_time = (rep_df["left_time_boundary"] + rep_df["right_time_boundary"]) / 2
rep_real_time = rep_time / mu * gen

plt.figure(figsize=(5, 6))

for t, r in zip(time_list, rccr_list):
    plt.plot(t, r, color='lightgray', alpha=0.2, linewidth=1, drawstyle='steps')

plt.plot(rep_real_time, rccr_mean, label="mean RCCR", linewidth=4)
plt.fill_between(rep_real_time, rccr_mean - rccr_std, rccr_mean + rccr_std,
                 alpha=0.3, label="± SD")

plt.axhline(y=0.5, color='black', linestyle='--', linewidth=3, alpha=0.7, label='RCCR = 0.5')

plt.xscale("log")
plt.xlim(5000,6.0e6)
plt.ylim(0, 1.05)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.xlabel("Years before the present", fontsize=22)
plt.ylabel("Relative cross coalescence rate", fontsize=20)
plt.grid(False)
plt.tight_layout()


# plt.savefig('msmc_rCCR_gen_2.5_mu_5.97e-9.svg')
# plt.savefig('msmc_rCCR_gen_2.5_mu_5.97e-9.pdf')
plt.show()
```

```python
def getCCRintersect(df, val):
    xVec = gen * ((df.left_time_boundary + df.right_time_boundary)/2) / mu
    yVec = 2.0 * df.lambda_01 / (df.lambda_00 + df.lambda_11)
    i = 0
    while yVec[i] < val:
        i += 1
    intersectDistance = (val - yVec[i - 1]) / (yVec[i] - yVec[i - 1])
    return xVec[i - 1] + intersectDistance * (xVec[i] - xVec[i - 1])
```

```python
files = glob.glob("msmc/combined_*.final.txt")

split_times = [getCCRintersect(pd.read_csv(f, sep="\t"), 0.5) for f in files]
print("Median split time:", np.median(split_times))
print("95% CI:", np.percentile(split_times, [2.5, 97.5]))
```
